This file contains examples of preliminary data visualizations. Some ideas include:
- The number of sponsors a bill has compared to the average
- Sankey chart for which steps bills make it to
- Number of bills and percent that pass for each primary sponsor
- Donation stats for each sponsor(big donations, corporate, percent in state)
- Summary stat visualizations for the homepage

```python
import pandas as pd
```

```python
# Donation stats, given a sponsor
sponsor = "Sara Feigenholtz"

df = pd.read_csv("../pull_IL_sunshine/intermediate_data/donation_stats.csv")
sponsor_row = df[df["Name"] == sponsor]

index = sponsor_row.index[0]

pd.DataFrame([sponsor_row["pct_c_IL_all"], 1-sponsor_row["pct_c_IL_all"]], 
    index=["In state donations", "Out of state donations"]).plot.pie(y= index, legend=False)

pd.DataFrame([sponsor_row["pct_$_non1a_all"], 1-sponsor_row["pct_$_non1a_all"]], 
    index=["Corporate donations ($)", "Individual donations ($)"]).plot.pie(y = index, legend=False)

pd.DataFrame([sponsor_row["pct_c_non1a_all"], 1-sponsor_row["pct_c_non1a_all"]], 
    index=["Corporate donations (#)", "Individual donations (#)"]).plot.pie(y = index, legend=False)

pd.DataFrame([sponsor_row["pct_c_above_all"], 1-sponsor_row["pct_c_above_all"]], 
    index=["Donations above $5,000 (#)", "Other donations (#)"]).plot.pie(y = index, legend=False)

pd.DataFrame([sponsor_row["pct_$_above_all"], 1-sponsor_row["pct_$_above_all"]], 
    index=["Donations above $5,000 ($)", "Other donations ($)"]).plot.pie(y = index, legend=False)
```
